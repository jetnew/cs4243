```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.cm as cm
```

# Load datasets

```python
dataset = np.load('dataset/dataset.npy', allow_pickle=True)[()]
dataset.keys()
```

```python
cifar_train = dataset['cifar_train']
cifar_test = dataset['cifar_test']
quickdraw = dataset['quickdraw']
sketchy_real = dataset['sketchy_real']
sketchy_doodle = dataset['sketchy_doodle']
tuberlin = dataset['tuberlin']
```

# Class Distributions

```python
def plot_class_dist(ax, d, title=""):
    ax.set_title(title)
    ax.bar([cat for cat in d.keys()], [x.shape[0] for x in d.values()])
```

```python
fig, ax = plt.subplots(2, 3, sharey='all', figsize=(15,6))
plt.yscale('log')
plot_class_dist(ax[0,0], cifar_train, "CIFAR-Train")
plot_class_dist(ax[0,1], cifar_test, "CIFAR-Test")
plot_class_dist(ax[0,2], quickdraw, "QuickDraw")
plot_class_dist(ax[1,0], sketchy_real, "Sketchy-Real")
plot_class_dist(ax[1,1], sketchy_doodle, "Sketchy-Doodle")
plot_class_dist(ax[1,2], tuberlin, "TUBerlin")
plt.tight_layout()
plt.show()
```

# Mean Image

```python
def plot_mean_imgs(d):
    cats = list(d.keys())
    col = 5 if len(cats) > 8 else 4
    fig, ax = plt.subplots(2, col, figsize=(15,6))
    plt.axis('off')
    for r in range(2):
        for c in range(col):
            try:
                mean_img = d[cats[r * col + c]].mean(axis=0).astype(int)
                ax[r,c].axis('off')
                ax[r,c].set_title(cats[r * col + c])
                ax[r,c].set_xticks([])
                ax[r,c].set_yticks([])
                ax[r,c].imshow(mean_img, cmap='gray')
            except:
                break
    plt.show()
```

```python
plot_mean_imgs(cifar_train)
```

```python
plot_mean_imgs(cifar_test)
```

```python
plot_mean_imgs(quickdraw)
```

```python
plot_mean_imgs(sketchy_real)
```

```python
plot_mean_imgs(sketchy_doodle)
```

```python
plot_mean_imgs(tuberlin)
```

# Dimensionality Reduction Plots

```python
def plot_tsne(d, title="", alpha=1, size=1):
    class_to_idx = {clas: idx for idx, clas in enumerate(d.keys())}
    idx_to_class = {idx: clas for idx, clas in enumerate(d.keys())}
    
    labels = []
    for clas, data in d.items():
        labels += [class_to_idx[clas]] * data.shape[0]
    labels = np.asarray(labels)
    
    data = np.concatenate(list(d.values()))
    data = data.reshape((data.shape[0], -1))
    pc = PCA(n_components=50).fit_transform(data)
    
    c = TSNE(n_components=2).fit_transform(pc)
    
    colors = cm.rainbow(np.linspace(0, 1, len(d.keys())))
    plt.title(title)
    for clas, color in zip(d.keys(), colors):
        idx = np.where(labels == class_to_idx[clas])[0]
        plt.scatter(c[idx][:,0], c[idx][:,1], label=clas, s=size, alpha=alpha, color=color)
    plt.legend(loc='best', markerscale=2)
    plt.show()
```

```python
plot_tsne(cifar_train, "CIFAR-Train")
```

```python
plot_tsne(cifar_test, "CIFAR-Test")
```

```python
plot_tsne(quickdraw, "QuickDraw")
```

```python
plot_tsne(sketchy_real, "Sketchy-Real", alpha=1, size=4)
```

```python
plot_tsne(sketchy_doodle, "Sketchy-Doodle", alpha=1, size=2)
```

```python
plot_tsne(tuberlin, "TUBerlin", alpha=1, size=4)
```
